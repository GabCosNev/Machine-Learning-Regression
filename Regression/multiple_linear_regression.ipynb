# Multiple Linear Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0) # Treinamento Seleção 20% para teste random definido
```

## Training the Multiple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression().fit(X_train, y_train) # Treinamento linear
```

## Predicting the Test set results

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2) # Printa apenas duas casas decimais para arrays
print(np.column_stack((y_pred, y_test))) # Juntando os dois arrays
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue')  # Plotando os pontos de previsão
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red')  # Linha de identidade
plt.title('Previsão vs Valores Reais')
plt.xlabel('Valores Reais')
plt.ylabel('Previsões')
plt.grid()
plt.show()
```

```python
AP_test = X_test[:, 1]
plt.figure(figsize=(10, 6))
plt.scatter(AP_test, y_pred, color='blue', label='Previsões')  # Plotando as previsões
plt.scatter(AP_test, y_test, color='orange', label='Valores Reais')  # Plotando os valores reais
plt.title('AP vs Previsões')
plt.xlabel('AP')
plt.ylabel('Previsões')
plt.legend()
plt.grid()
plt.show()
```

## Evaluating the Model Performance

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
