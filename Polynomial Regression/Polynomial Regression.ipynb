```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_freature = PolynomialFeatures(degree=4) # Select n from y = b0 + b1x1 + b2x1**2 + bnx1**n
X_poly = poly_freature.fit_transform(X)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, y)
print(X_poly)
```

```python
plt.figure(figsize=(10, 6))                                #Transparência
plt.scatter(X, y, color = 'red', alpha=0.6, label='Level')
plt.plot(X, lin_reg.predict(X), color = 'blue', label='Salary')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Truth or Bluff')
plt.legend() # Legenda em cima
plt.grid() # Linhas do gráfico
plt.show() # mostrar gráfico
```

```python
plt.figure(figsize=(10, 6))                                #Transparência
plt.scatter(X, y, color = 'red', alpha=0.6, label='Level')
plt.plot(X, lin_reg2.predict(X_poly), color = 'blue', label='Salary')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Truth or Bluff')
plt.legend() # Legenda em cima
plt.grid() # Linhas do gráfico
plt.show() # mostrar gráfico
```

```python
# Perever o salário na posição nivel 6.5 Linear Regression
print(lin_reg.predict([[6.5]])) # Primeiro colchete é as linhas o segundo é as colunas
```

```python
# Perever o salário na posição nivel 6.5 )Polymiar Regression
print(lin_reg2.predict(poly_freature.transform([[6.5]])))
```
